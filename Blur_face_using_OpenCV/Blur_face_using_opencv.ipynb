# **second code for blur face**

```python
# importing libraries
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
  
image = cv2.imread('/content/drive/My Drive/Colab Notebooks/my projects/blur_face_opencv/download.jpg')
  
cv2_imshow(image)
cv2.waitKey(0)
```

```python
# Gaussian Blur
Gaussian = cv2.GaussianBlur(image, (21, 21), 0)
cv2_imshow(Gaussian)
cv2.waitKey(0)
```

```python
# Median Blur
median = cv2.medianBlur(image, 15)
cv2_imshow(median)
cv2.waitKey(0)
```

```python
#A bilateral filter is a non-linear, edge-preserving, and noise-reducing smoothing filter for images.
# Bilateral Blur
bilateral = cv2.bilateralFilter(image, 9, 75, 75)
cv2_imshow(bilateral)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# **Blur face in vedio**

```python
import cv2
import numpy as np
def resize(img,new_width=500):
    height,width,_ = img.shape
    ratio = height/width
    new_height = int(ratio*new_width)
    return cv2.resize(img,(new_width,new_height))
```

```python
face_cascade = cv2.CascadeClassifier("/content/drive/My Drive/Colab Notebooks/my projects/blur_face_opencv/haarcascade_frontalface_alt.xml")
cap = cv2.VideoCapture("/content/drive/My Drive/Colab Notebooks/my projects/blur_face_opencv/v1.mp4")
```

```python
while True:
    _,frame = cap.read()
    frame = resize(frame)
    detections = face_cascade.detectMultiScale(frame,scaleFactor=1.1,minNeighbors=6)
    for face in detections:
        x,y,w,h = face
        frame[y:y+h,x:x+w] = cv2.GaussianBlur(frame[y:y+h,x:x+w],(15,15),cv2.BORDER_DEFAULT)
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        cv2_imshow(frame)
        if cv2.waitKey(1) == 27:
            break
cap.release()
cv2.destroyAllWindows()
```
